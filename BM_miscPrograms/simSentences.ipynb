```python
import spacy
import docx
import docx2txt

from nltk.tokenize import word_tokenize
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
text = docx2txt.process("BS2.docx")
```

```python
def returnRedundantSentences(document, threshold):
    
    document = nlp(document)
    
    redundantSentences = []
    
    for sentenceA in document.sents:
        
        rs = []
        
        for sentenceB in document.sents:
            if sentenceA.similarity(sentenceB) > threshold and sentenceA.similarity(sentenceB) < 1.0:
                rs.append(sentenceB)
        if rs:
            redundantSentences.append([sentenceA.text.upper()] + [r.text for r in rs])
    return redundantSentences
```

```python
RedundantSentences = returnRedundantSentences(text, 0.9)
```

```python
returnRedundantSentences(text, 0.9)
```

```python
for i in RedundantSentences:
    print("Sentence set: \n")
    for j in i:
        print(j)
    print("*****************************************************************************************")
```
