```python
import docx
from pprint import pprint
import nltk
nltk.download('stopwords')
```

```python
from Questgen import main
qe= main.BoolQGen()
```

```python
doc = docx.Document('BS2.docx')
```

```python
text_list = []

for paragraph in doc.paragraphs:
    text_list.append(paragraph.text.strip())
```

```python
dict_list = [{'input_text': text} for text in text_list]
```

```python
# Boolean questions - potential to use as an objective list and as titles in the formatted draft
for dict in dict_list:
    output = qe.predict_boolq(dict)
    pprint (output)
```

```python
# Multiple choice questions - append to end of draft
for dict in dict_list:
    qg = main.QGen()
    output = qg.predict_mcq(dict)
    pprint (output)
```

```python
# Generate FAQs - also potential for section titles or objectives in the formatted draft
for dict in dict_list:
    output = qg.predict_shortq(dict)
    pprint (output)
```
